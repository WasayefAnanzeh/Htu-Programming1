```
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import pandas as pd
df = pd.read_csv('covid_19_data.csv')


countries = df.groupby(by=['Country/Region'], as_index=False).sum()
countries['Active'] = countries['Confirmed'] - countries['Deaths'] - countries['Recovered']

countries = countries.sort_values(by='Confirmed', ascending=False)
countries = countries.head(10)

country_names = countries['Country/Region']

counts_confirmed = countries['Confirmed']
counts_death = countries['Deaths']
countries['active_cases'] = countries['Confirmed'] - countries['Deaths']-countries['Recovered']
```

```
import matplotlib.pyplot as plt 

deaths = counts_death.tail (5)
labels = country_names.tail(5)
size  = [15,30,45,10]
explode = (0,0.1,0,0,0)

figl , ax1 = plt.subplots()
ax1.pie(deaths , explode=explode , labels = labels , autopct = '%1.1f%%')
ax1.axis('equal')
plt.show

```

```
figure (num=None , figsize=(12,6) , dpi=80,facecolor='w' , edgecolor='k')

plt.scatter(counts_death,counts_confirmed , color='red')
plt.ylable('counts')
plt.title('corona stats')

plt.show()
```
